```
# As sysBP and diaBP are highly correlational, one of them can be dropped.
df.drop(['sysBP'],axis=1,inplace=True)
```

```
df['TenYearCHD']=df['TenYearCHD'].astype("int")
```

```
x=df.drop('TenYearCHD',axis=1)
y=df['TenYearCHD']
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```
# Applying Naive-Bayes Classifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```
NB=GaussianNB()
NB.fit(x_train,y_train)
```

```
y_train_predict=NB.predict(x_train)
model_score=NB.score(x_train,y_train)
print(model_score)
print(metrics.confusion_matrix(y_train,y_train_predict))
```

```
y_test.value_counts()
```

```
y_test_predict=NB.predict(x_test)
```

```
score_model=NB.score(x_test,y_test)
print(score_model)
print(metrics.confusion_matrix(y_test,y_test_predict))
```

```
# Applying KNN Model
from sklearn.neighbors import KNeighborsClassifier
KNN_model=KNeighborsClassifier()
KNN_model.fit(x_train,y_train)
```

```
y_train_pred=KNN_model.predict(x_train)
model_score1=KNN_model.score(x_train,y_train)
print(model_score1)
print(metrics.confusion_matrix(y_train,y_train_pred))
print(metrics.classification_report(y_train,y_train_pred))
```
